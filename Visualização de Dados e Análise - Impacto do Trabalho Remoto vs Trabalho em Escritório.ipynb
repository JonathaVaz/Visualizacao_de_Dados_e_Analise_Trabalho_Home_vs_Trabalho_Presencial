### **Importando Bibliotecas para o Desenvolvimento do Projeto**

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

### **Carregando o Arquivo e Visualizando Brevemente**

```python
df = pd.read_csv('The Impacts of Working Remotely and in an Office Survey.csv')
```

```python
df.head(5)
```

### **Verificando a Estatística Geral do Arquivo**

```python
df.info()
```

```python
df.describe()
```

### **Criando Variável para Cada Coluna e Suguir para Visualização**

```python
experimentou_trabalhar_em_casa = df['Have you ever experienced working from home?'].value_counts().sort_index()  # Você já experimentou trabalhar em casa?

aumento_de_produtividade = df['Do you think that working from home increases your work productivity?'].value_counts().sort_index() # Você acha que trabalhar em casa aumenta sua produtividade no trabalho?

trabalhar_em_casa_impede_de_sair = df['Do you think that working from home prevents you from going out?'].value_counts().sort_index()  # Você acha que trabalhar em casa te impede de sair?

mais_flexibilidade_trabalhando_em_casa = df['Do you think that working from home gives you more flexibility?'].value_counts().sort_index() # Você acha que trabalhar em casa lhe dá mais flexibilidade?

economizar_mais_tempo = df['Do you think that working from home saves you more time?'].value_counts().sort_index()  # Você acha que trabalhar em casa economiza mais tempo?

potencial_problema_fisico = df['Which work type has the potential to cause you physical problems, such as the neck, back, and leg pain?'].value_counts().sort_index()  # Que tipo de trabalho tem potencial para causar problemas físicos, como dores no pescoço, nas costas e nas pernas?

potencial_problema_psicologico = df['Which work type has the potential to cause mental disorders, such as stress, anxiety, and depression?'].value_counts().sort_index()  # Que tipo de trabalho tem potencial para causar transtornos mentais, como estresse, ansiedade e depressão?

impedir_entrar_em_contato = df['Do you think that working from home prevents you from getting in contact with people?'].value_counts().sort_index()  # Você acha que trabalhar em casa impede você de entrar em contato com as pessoas?

tipo_foco_trabalho = df['Which work type keeps you focused while working?'].value_counts().sort_index()  # Que tipo de trabalho mantém você focado enquanto trabalha?

trabalho_preferido = df['Which work type do you prefer the most?'].value_counts().sort_index()  # Qual tipo de trabalho você mais prefere?
```

## **Iniciando Processo de Visualização de Dados**

### **Traçando Gráficos de Barras e Circular (Setores ou Pizza) para Cada Pergunta**

```python
# Definindo Paleta de Cores e Estilo do Gráfico
plt.style.use('dark_background')
paleta_duo = ['mediumblue', 'blueviolet']
paleta_tripla = ['deeppink', 'mediumblue', 'blueviolet']
```

### **Análise da Distribuição de Amostras se o Usuário já experimentou trabalhar em casa**

```python
plt.figure(figsize=(15,11))
plt.suptitle('Distribuição de Amostras se o Usuário já Experimentou Trabalhar em Casa\n', fontsize=17, fontweight='bold', style='italic')
plt.subplot(2, 2, 1)
plt.bar(experimentou_trabalhar_em_casa.index, experimentou_trabalhar_em_casa.values, edgecolor='white', color=paleta_duo)
plt.grid(alpha=0.2)
plt.xlabel('\nRespostas', fontsize=12, fontweight='bold', style='italic') 
plt.ylabel('Contagem', fontsize=11, fontweight='bold', style='italic')
plt.xticks(fontweight='bold', style='italic')
plt.yticks(fontweight='bold', style='italic')
plt.title('Distribuições das Respostas\n', fontsize=12, fontweight='bold', style='italic') 


plt.subplot(2, 2, 2)
plt.pie(experimentou_trabalhar_em_casa.values, labels=experimentou_trabalhar_em_casa.index, autopct='%1.1f%%', colors=paleta_duo)
plt.title('Percentual de Respostas\n', fontsize=12, fontweight='bold', style='italic') 
```

```python
plt.figure(figsize=(15,11))
plt.suptitle('Distribuição de Amostras se o Usuário acha que Trabalhar em Casa Aumenta sua Produtividade no Trabalho\n', fontsize=15, fontweight='bold', style='italic')
plt.subplot(2, 2, 1)
plt.bar(aumento_de_produtividade.index, aumento_de_produtividade.values, edgecolor='white', color=paleta_duo)
plt.grid(alpha=0.2)
plt.xlabel('\nRespostas', fontsize=12, fontweight='bold', style='italic') 
plt.ylabel('Contagem', fontsize=11, fontweight='bold', style='italic')
plt.xticks(fontweight='bold', style='italic')
plt.yticks(fontweight='bold', style='italic')
plt.title('Distribuições das Respostas\n', fontsize=12, fontweight='bold', style='italic') 


plt.subplot(2, 2, 2)
plt.pie(aumento_de_produtividade.values, labels=aumento_de_produtividade.index, autopct='%1.1f%%', colors=paleta_duo)
plt.title('Percentual de Respostas\n', fontsize=12, fontweight='bold', style='italic') 
```

## ***Conclusão da Análise:***
### **- No Decorrer da Análise conseguimos Observar que Existe um Percentual Equilibrado e nada Anormal de Pessoas que já Experimentaram Trabalhar em Home Office, o que Corresponde a 60% dos Usuários.**
### **- A Análise também aponta que a Produtividade Aumenta de acordo com os Usuários um Número Considerável de Respostas Positivas (70%).**
### **- Em Relação ao Trabalho em Casa Impede de Sair, 70% das Respostas foram de acordo, enquanto 30% desacreditam.**
### **- E 100% dos Usuários Acreditam que Trabalhar em Casa lhe Permitem mais Flexibilidade, um Valor Surpreendente e Minuncioso.**
### **- Apenas 40% Acreditam que Trabalhar em Home Office Economiza mais Tempo, enquanto 60% Não concordam. São Valores Consideravelmente Equilibrados em Relação a Análise.**
### **- Um Fator Importante Presente na Análise é que 80% dos Usuários Acreditam que o Trabalho em Casa Geram Problemas Fisicos, como Dores no Pescoço, na Costas, e nas Pernas. Isso é Algo de Extrema Importância de ser Informado, para Futuras Medidas de Solução.**
### **- 70% dos Usuários Responderam que o Trabalho em Casa Geram Transtornos Mentais, tais como, Estresse Psicologico, Depressão e Ansiedade. Outro Fator Importante de ser Análisado.**
### **- O Trabalho em Home Office de acordo com os 70% dos Usuários te Impede de Entrar em Contato com outras Pessoas, é uma Avaliação Normal e Aceitável, por outro lado 30% acreditam que não.**
### **- Trabalhar no Escritório ou Presencialmente faz 60% dos Usuários ficar mais Concentrados em seu Trabalho de acordo com os mesmos, já 40% acham que o Trabalho em Casa o fazem ficar concentrados em suas atividades.**
### **- Por Último e Talvez o mais Importante, Qual o Tipo de Trabalho que mais Preferem? Bom... de Acordo com as Informações Concebidas, 50% dos Usuários Preferem o Trabalho Híbrido, 30% Preferem Mais o Trabalho Presencial e 20% dos Usuários Preferem o Modelo Home Office. Três Fatores Importântes da Análise.**
